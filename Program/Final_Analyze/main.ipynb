```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV 檔案
filename = "/Users/corn/Documents/Program/2025_Rocket/Test_Record/Final/ground_data.csv"
df = pd.read_csv(filename)
```

```python
# 輸入兩個欄位名
x_col = "slope_data_0"
y_col = "slope_data_1"

# 繪圖
plt.figure(figsize=(10, 6))
plt.plot(df[x_col], df[y_col], linestyle='-')  # 可以改成 scatter() 看起來會像散點圖
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title(f"{y_col} vs {x_col}")
plt.grid(True)
plt.tight_layout()
plt.show()
```
