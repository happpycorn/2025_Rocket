```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 顏色變數
color_alt = '#019a7b'
color_slope = '#1f318f'

# 讀取 CSV
df = pd.read_csv(r'C:\Users\happp\OneDrive\文件\2025_Rocket\Program\Elevator_Test\data-250409.csv')
```

```python
# 畫圖
fig, ax1 = plt.subplots(dpi=300)

# 左邊 y 軸：average_alt
ax1.set_xlabel('Time(ms)')
ax1.set_ylabel('Average Altitude(m)', color=color_alt)
ax1.plot(df["Time"], df["average_alt"], color=color_alt, label='Average Altitude')
ax1.tick_params(axis='y', labelcolor=color_alt)
ax1.set_ylim(-20, 60)

# 加入左邊 0 線
ax1.axhline(0, color='gray', linestyle='--', linewidth=1, alpha=0.6)

# 右邊 y 軸：slope
ax2 = ax1.twinx()
ax2.set_ylabel('Slope(m/s)', color=color_slope)
ax2.plot(df["Time"], df["slope"], color=color_slope, label='Slope')
ax2.tick_params(axis='y', labelcolor=color_slope)
ax2.set_ylim(-3, 9)

# 加入右邊 0 線
ax2.axhline(0, color='gray', linestyle='--', linewidth=1, alpha=0.6)

# 顯示圖
plt.title("Average Altitude and Slope Over Time")
fig.tight_layout()
plt.show()
```

```python
# 顏色設定
color_alt1 = '#019a7b'
color_alt2 = '#1f318f'
color_avg = 'black'
```

```python
# 畫圖
fig, ax = plt.subplots(dpi=300)

# 繪製 alt_1、alt_2、average_alt
ax.plot(df["Time"], df["average_alt"], label="Average Alt", linestyle='-', color=color_avg)
ax.plot(df["Time"], df["alt_1"], label="Alt 1", linestyle='--', color=color_alt1)
ax.plot(df["Time"], df["alt_2"], label="Alt 2", linestyle='--', color=color_alt2)

# 標題與標籤
ax.set_xlabel("Time(ms)")
ax.set_ylabel("Altitude(m)")
ax.set_title("Alt 1, Alt 2, and Average Altitude Over Time")

# 顯示圖例
ax.legend()

# 加上下邊界 padding
alt_min = min(df["alt_1"].min(), df["alt_2"].min(), df["average_alt"].min())
alt_max = max(df["alt_1"].max(), df["alt_2"].max(), df["average_alt"].max())
padding = (alt_max - alt_min) * 0.1
ax.set_ylim(alt_min - padding, alt_max + padding)

# 顯示圖
plt.tight_layout()
plt.show()
```
