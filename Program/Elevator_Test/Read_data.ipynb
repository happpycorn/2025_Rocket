```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 顏色變數
color_alt = '#019a7b'
color_slope = '#1f318f'

# 讀取 CSV
df = pd.read_csv(r'C:\Users\happp\OneDrive\文件\2025_Rocket\Program\Elevator_Test\data-250409.csv')
```

```python
# 畫圖
fig, ax1 = plt.subplots(dpi=300)

# 左邊 y 軸：average_alt
ax1.set_xlabel('Time(ms)')
ax1.set_ylabel('Average Altitude(m)', color=color_alt)
ax1.plot(df["Time"], df["average_alt"], color=color_alt, label='Average Altitude')
ax1.tick_params(axis='y', labelcolor=color_alt)
ax1.set_ylim(-20, 60)

# 加入左邊 0 線
ax1.axhline(0, color='gray', linestyle='--', linewidth=1, alpha=0.6)

# 右邊 y 軸：slope
ax2 = ax1.twinx()
ax2.set_ylabel('Slope(m/s)', color=color_slope)
ax2.plot(df["Time"], df["slope"], color=color_slope, label='Slope')
ax2.tick_params(axis='y', labelcolor=color_slope)
ax2.set_ylim(-3, 9)

# 加入右邊 0 線
ax2.axhline(0, color='gray', linestyle='--', linewidth=1, alpha=0.6)

# 顯示圖
plt.title("Average Altitude and Slope Over Time")
fig.tight_layout()
plt.show()
```
