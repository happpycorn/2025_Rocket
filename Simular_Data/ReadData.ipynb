### Draw

```python
plt.plot(data['Vertical Speed (m/s)'], data['High (km)'])
```

```python
color1 = (0.122, 0.192, 0.561)
color2 = (0.004, 0.604, 0.482)

# 創建圖表和第一個 y 軸
fig, ax1 = plt.subplots()

# 第一組數據：高度（High）
ax1.plot(data['Time (s)'], data['High (km)'], color=color1, label='High (km)')
ax1.axhline(0, color="black", linestyle="--", linewidth=1)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('High (km)')
ax1.set_ylim(-0.216, 1.3)
ax1.tick_params(axis='y', labelcolor=color1)

# 創建第二個 y 軸
ax2 = ax1.twinx()

# 第二組數據：垂直速度（Vertical Speed）
ax2.plot(data['Time (s)'], data['Vertical Speed (m/s)'], color=color2, label='Vertical Speed (m/s)', linestyle='--')
ax2.set_ylabel('Vertical Speed (m/s)')
ax2.set_ylim(-30, 180)
ax2.tick_params(axis='y', labelcolor=color2)

# 添加標題
plt.title('High vs Vertical Speed over Time')

# 調整佈局避免重疊
fig.tight_layout()

plt.savefig('output_chart.png', dpi=300, bbox_inches='tight')  # 儲存為 PNG 格式

# 顯示圖表
plt.show()
```
